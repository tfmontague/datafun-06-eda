### Step 5: Data Transformation and Feature Engineering
Enhance the dataset with additional features or transformations that might help in the analysis. To accomplish this, we will categorize the planets based on their size (mass) and distance from Earth.

```python
# Function to categorize planets based on size
def categorize_planet_size(mass):
    if mass <= 0.03:  # 10 Earth masses
        return 'Earth-like'
    elif mass <= 0.1:  # >10 and 30 Earth masses
        return 'Super-Earth'
    elif mass <= 0.5:  # >30 and 160 Earth masses
        return 'Neptune-like'
    else:  # >160 Earth masses
        return 'Jupiter-like'

# Apply the function to create a new column for planet size category
planets['size_category'] = planets['mass'].apply(categorize_planet_size)

# Check the first few rows to verify the new column
print(planets[['mass', 'size_category']].head())
```

```python
# Function to categorize planets based on distance from Earth
def categorize_planet_distance(distance):
    if distance <= 20:
        return 'Close Proximity'
    elif distance <= 100:
        return 'Moderately Distant'
    elif distance <= 500:
        return 'Distant'
    else:
        return 'Far Away'

# Apply the function to create a new column for planet distance category
planets['distance_category'] = planets['distance'].apply(categorize_planet_distance)

# Check the first few rows to verify the new column
print(planets[['distance', 'distance_category']].head())
```

```python
# Function to transform planet mass to a suitable size for bubble visualization
def transform_mass_to_size(mass):
    # Apply a non-linear transformation - square root scaling
    # Adjust the scaling factor as needed to fit the visual context of the plot
    return np.sqrt(mass) * 10  # The constant factor can be adjusted for visual preference

# Apply the transformation to the mass column to create a size column
planets['size'] = planets['mass'].apply(transform_mass_to_size)

# Verify the transformation with the first few rows
print(planets[['mass', 'size']].head())

```

### Step 6: Visual Analysis
Use various tools to visualize relationships between features and identify correlations or patterns.

```python
# Creating the Bar Chart to visualize the count of exoplanets categorized by their sizes: Jupiter-like, Neptune-like, Earth-like, and Super-Earth.
sns.countplot(x='size_category', data=planets, palette='coolwarm')
plt.title('Distribution of Planet Sizes')
plt.show()
```

Key Observations:
- **Dominance of Jupiter-like Planets:** The bar for Jupiter-like planets is significantly taller than the others, suggesting that most of the planets in this dataset are of this size category. This may indicate a bias in detection methods towards larger planets, as their size and mass can make them easier to detect through methods like radial velocity or transit observations.

- **Scarcity of Smaller Planets:** Earth-like and Super-Earth categories have significantly fewer counts compared to Jupiter-like planets. This might suggest that it is more challenging to detect smaller planets with current technology, or it could reflect a real frequency in the universe  that larger gas giants are more common than smaller rocky planets.

- **Technological Implications:** The predominance of larger planets in the dataset might reflect the era when these planets were discovered. Earlier in exoplanet research, detection methods were more likely to find larger planets. As technology and methods improve, we would expect to see the numbers for Earth-like and Super-Earth categories increase.

```python
# Creating the bar chart to show the counts of exoplanets categorized by their distance from Earth into four groups: Moderately Distant, Close Proximity, Distant, and Far Away the distribution of distance categories
sns.countplot(x='distance_category', data=planets, palette='viridis')
plt.title('Distribution of Planet Distance Categories')
plt.show()
```

Key Observations:

```python
# Creating the scatter plot graphing orbital period versus mass for different exoplanets, with size and color coding to represent different distance and size categories
sns.scatterplot(x='orbital_period', y='mass', hue='distance_category', style='size_category', data=planets, palette='magma', alpha=0.7)
plt.xscale('log')
plt.yscale('log')
plt.title('Orbital Period vs. Mass by Distance and Size Category')
plt.xlabel('Orbital Period (days)')
plt.ylabel('Mass (Jupiter Masses)')
plt.show()
```

```python
# Interactive 3D scatter plot to illustrate the relationships between mass, orbital period, and distance with size categories and distance as color-coded dimensions
import plotly.express as px

fig = px.scatter_3d(planets.dropna(), x='mass', y='orbital_period', z='distance',
                    color='size_category', symbol='distance_category',
                    title="3D Scatter Plot of Planets by Size and Distance Category")

# Establish the figure layout size
fig.update_layout(width=1300, height=1000)

fig.show()
```

Scatter Plot Key Observations: *(note-3D scatter plot displayed to support interactity)*

This is consistent with observational biases where larger planets, which have a more substantial gravitational pull and often cause more noticeable wobbles in their host stars, are easier to detect at greater distances.

This is typical of the planets detected by transit methods, where repeated passes in front of their host stars make them easier to observe.

- **Rare Massive Planets Close to Earth:** If there are only a few large bubbles close to the zero-point of the distance axis, it indicates that massive planets are rarely detected close to Earth, possibly due to a smaller observational volume or the fact that such systems are intrinsically rare.

- **Potential Detection Gaps:** Areas in the chart with sparse or no points could indicate regions where current detection methods are less effective, revealing potential gaps in our detection capabilities or in the actual distribution of exoplanets.

- **Spread of Planetary Masses:** A wide range of colors across the bubbles signifies a diverse spread of planetary masses within the observed exoplanet population.

**This stacked bar chart summarizes of the state of exoplanet discoveries, showcasing the types of planets that are most commonly detected and hinting at the biases inherent in our detection capabilities.**

- **Jupiter-like Dominance:** The most substantial segments in nearly all distance categories are those representing Jupiter-like planets, underscoring their prevalence in detections and suggesting that current detection methods are most sensitive to these large planets.

- **Fewer Smaller Planets Detected:** The smaller segments for Earth-like and Super-Earth planets, especially at greater distances, highlight the relative difficulty in detecting smaller planets with existing technology.

- **Distribution Across Distances:** The distribution of planets across the distance categories shows that a significant number of planets are detected in the 'Moderately Distant' category, indicating the capability of current methods to find planets over a range of distances.

- **Close Proximity Observations:** The 'Close Proximity' category shows a more diverse size distribution of planets, albeit with fewer overall detections, which may reflect either a true lower occurrence rate or limitations in the volume of space surveyed.

```python
# Creating stacked bar chart to summarize the state of exoplanet discoveries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
planets_data = pd.read_csv('data/planets.csv')

# Function to categorize planet size
def categorize_size(mass):
    if pd.isnull(mass):
        return np.nan
    elif mass <= 0.03:
        return 'Earth-like'
    elif mass <= 0.1:
        return 'Super-Earth'
    elif mass <= 0.5:
        return 'Neptune-like'
    else:
        return 'Jupiter-like'

# Function to categorize distance
def categorize_distance(distance):
    if pd.isnull(distance):
        return np.nan
    elif distance <= 20:
        return 'Close Proximity'
    elif distance <= 200:
        return 'Moderately Distant'
    elif distance <= 1000:
        return 'Distant'
    else:
        return 'Far Away'

# Apply the categorization functions
planets_data['size_category'] = planets_data['mass'].apply(categorize_size)
planets_data['distance_category'] = planets_data['distance'].apply(categorize_distance)

# Create a pivot table for the count of planets by size and distance category
planet_counts = planets_data.pivot_table(index='distance_category', columns='size_category', aggfunc='size', fill_value=0)

# Plotting
planet_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Number of Planets by Size and Distance Category')
plt.xlabel('Distance Category')
plt.ylabel('Number of Planets')
plt.legend(title='Size Category')
plt.tight_layout()
plt.show()

```

### Step 8: Conclusion
Our EDA underscores the dynamic nature of exoplanet discovery, heavily influenced by technological capabilities and the methods employed. While gas giants have dominated past discoveries, there's a clear pathway for the increased detection of smaller, potentially Earth-like planets, as our technological and methodological approaches advance. 

As indicated in linear regression analysis below, the optimistic projection for future discoveries not only reflects the potential for technological evolution but also highlights the growing capability of astronomers to peer deeper into our galaxy, uncovering the secrets of distant worlds. This journey of discovery is far from complete, and the coming years promise to be a golden era for uncovering new worlds across the cosmos.


```python
# Linear regression analysis to project the potential for future discoveries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd

# Load the dataset 
planets_data = pd.read_csv('data/planets.csv')

# For the prediction, we will use data from 1989 to 2011
# Aggregate the total number of planets discovered per year
yearly_totals = planets_data.groupby('year').size()

# Filter data to include years up to 2011
yearly_totals_up_to_2011 = yearly_totals.loc[yearly_totals.index <= 2011]

# Fit a simple linear regression to this data to get a slope
X_up_to_2011 = yearly_totals_up_to_2011.index.values.reshape(-1, 1)  # Years (as independent variable)
y_up_to_2011 = yearly_totals_up_to_2011.values  # Total number of planets (as dependent variable)

# Create and fit the model
model_up_to_2011 = LinearRegression()
model_up_to_2011.fit(X_up_to_2011, y_up_to_2011)

# Predict the future based on the trend from 1989 to 2011
future_years_up_to_2011 = np.arange(X_up_to_2011[-1] + 1, X_up_to_2011[-1] + 21).reshape(-1, 1) # 20 years into the future
future_predictions_up_to_2011 = model_up_to_2011.predict(future_years_up_to_2011)

# Create the subplot for future predictions based on data from 1989 to 2011
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the historical data up to 2011
ax.plot(yearly_totals_up_to_2011.index, yearly_totals_up_to_2011.values, label='Historical Data (up to 2011)', marker='o')

# Plot the extrapolated future data
ax.plot(future_years_up_to_2011.flatten(), future_predictions_up_to_2011, label='Future Predictions (Extrapolated from 1989-2011)', marker='o', linestyle='--')

ax.set_title('Future Predictions of Planet Discoveries (Based on 1989-2011 Trend)')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Planets Discovered')
ax.legend()

plt.tight_layout()
plt.show()

```
